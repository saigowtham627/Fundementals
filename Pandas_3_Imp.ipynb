```python
points_df = pd.DataFrame(
    {
    '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
    '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)
points_df
```

```python
def eucldiean_distance(row): #Row is the parameter here, which has 1st point and 2nd point
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['ed'] = points_df.apply(eucldiean_distance, axis = 1) #axis must be specified as we are sending the rows but not the columns
```

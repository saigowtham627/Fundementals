```python
team_1 = input('Enter Team_1: ')
team_2 = input('Enter Team_2: ')

def custom(team_1, team_2):
    # Filter for matches where the two teams played against each other
    result = ipl[((ipl['Team1'] == team_1) & (ipl['Team2'] == team_2)) |
                 ((ipl['Team1'] == team_2) & (ipl['Team2'] == team_1))]

    # Get the count of wins for each team
    win_counts = result['WinningTeam'].value_counts()

    return win_counts

# Call the function and print the result
track_record = custom(team_1, team_2)
print(track_record)
```
